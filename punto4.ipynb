```python
import numpy as np
import matplotlib.pyplot as plt
```

# Punto 4

$x=(x_1,x_2), y=(y_1,y_2) \in \mathbb{R^2}$

$d(x,y) = \begin{cases} |x_2 - y_2| & x_1 = y_1 \\ |x_1-y_1|+|x_2|+|y_2| & x_1 \ne y_1 \end{cases}$

```python
data = np.random.randn(10000,2)
[filas, columnas] = data.shape
media = np.mean(data,0) #punto y
metrica = []
for j in range(0, filas):
    dif = data[j,:]-media
    if dif[0]==media[1]:
        d = abs(dif[1]-media[1])
        metrica.append(d)
    else: 
        d = abs(dif[0]-media[1])+abs(dif[1])+abs(media[1])
        metrica.append(d)
```

```python
p90 = np.percentile(metrica, 90, interpolation='linear')
I = np.argwhere(metrica > p90)
plt.plot(data[:,0], data[:,1],'ob')
plt.plot(data[I,0], data[I,1], 'or')
plt.show
```

$x=(x_1,x_2), y=(y_1,y_2) \in \mathbb{R^2}$

$d(x,y) = \begin{cases} \sqrt{(x_1-y_1)^2+(x_2-y_2)^2} & x_1^2+x_2^2 = y_1^2+y_2^2 \\ \sqrt{x_1^2+x_2^2}+\sqrt{y_1^2+y_2^2} & x_1^2+x_2^2 \ne y_1^2+y_2^2 \end{cases}$

```python
data = np.random.randn(10000, 2)
filas, columnas = data.shape
media = np.mean(data,0) #punto y
metrica = []
for j in range(0, filas-1):
    dif = data[j,:] - media 
    if (dif[0]**2 + media[0]**2) == (dif[1]**2 + media[1]**2):
        d = np.sqrt((dif[0]-dif[1])**2+(media[0]-media[1])**2)
        metrica.append(d)
    else:
        d = np.sqrt(dif[0]**2+media[0]**2) + np.sqrt(dif[1]**2+media[1]**2)
        metrica.append(d)
```

```python
p90 = np.percentile(metrica, 90, interpolation='linear')
I = np.argwhere(metrica > p90)
plt.plot(data[:,0], data[:,1],'ob')
plt.plot(data[I,0], data[I,1], 'or')
plt.show
```
