```python
from skimage.transform import resize
from skimage import io
from skimage.color import rgb2gray
import matplotlib.pyplot as plt
import os
import numpy as np
from numpy import linalg as LA
```

```python
path = os.listdir('imagenes')
imagenes = {}
for file in path:
    if file =='.DS_Store' or file == 'fondo.JPG':
        continue
    img = io.imread('imagenes/'+file,as_gray=True)
    imagenes[file[:2]] = resize(img, (700, 500), anti_aliasing=True)
```

```python
norm1 = {}
norm2 = {}
normInf = {}
normFro = {}
nati = {}
```

```python
for k,v in imagenes.items():
    if 'N' in k:
        nati[k] = v

for ki,vi in nati.items():
    for kj, vj in imagenes.items():
        if ('N' in kj): continue
        norm1[f'{ki}_{kj}'] = LA.norm( vi - vj,1)
        norm2[f'{ki}_{kj}'] = LA.norm(vi - vj,2)
        normInf[f'{ki}_{kj}'] = LA.norm(vi - vj, np.inf)
        normFro[f'{ki}_{kj}'] =  LA.norm(vi - vj, 'fro')
```

```python
print(sorted(norm1, key=norm1.get))
print(sorted(norm2, key=norm2.get))
print(sorted(normInf, key=normInf.get))
print(sorted(normFro, key=normFro.get))
```

```python
for kj, vj in imagenes.items():
    norm1[f'{kj}'] = LA.norm( imagenes['N1'] - vj,1)
    norm2[f'{kj}'] = LA.norm(imagenes['N1'] - vj,2)
    normInf[f'{kj}'] = LA.norm(imagenes['N1'] - vj, np.inf)
    normFro[f'{kj}'] =  LA.norm(imagenes['N1'] - vj, 'fro')
```

```python
print(sorted(norm1, key=norm1.get))
print(sorted(norm2, key=norm2.get))
print(sorted(normInf, key=normInf.get))
print(sorted(normFro, key=normFro.get))

print(norm1['D6'])
print(norm2['A1'])
print(normInf['C4'])
print(normFro['C1'])
```
